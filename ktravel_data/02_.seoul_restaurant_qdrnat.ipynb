```python
import pandas as pd
from langchain_core.documents import Document

# CSV_FILE ë³€ìˆ˜ì™€ df (DataFrame) ë³€ìˆ˜ëŠ” ì™¸ë¶€ì—ì„œ ì •ì˜ë˜ì—ˆë‹¤ê³  ê°€ì •í•©ë‹ˆë‹¤.

print("ğŸ“‚ CSV íŒŒì¼ ë¡œë”© ë° ì „ì²˜ë¦¬ ì¤‘...")

# ì‚¬ìš©ìê°€ ì œê³µí•œ ë ˆìŠ¤í† ë‘ ì»¬ëŸ¼ ëª©ë¡:
# ['restaurant_id', 'restaurant_name', 'image_path', 'Latitude', 
#  'Longitude', 'restaurant_name_en', 'place_en', 'near_subway_en', 
#  'type_en', 'description_clean_en']

print(f"âœ… ì´ {len(df)}ê°œ í•­ëª© ë¡œë“œ")
print(f"\nì»¬ëŸ¼: {df.columns.tolist()}")

# ì˜ë¯¸ ìˆëŠ” Document ìƒì„±
docs = []
for idx, row in df.iterrows():
    # 1. ì„ë² ë”©í•  í…ìŠ¤íŠ¸ (page_content) ìƒì„± - ê²€ìƒ‰ ì„±ëŠ¥ í–¥ìƒì„ ìœ„í•´ ì¤‘ìš”í•œ ì •ë³´ë§Œ ì¡°í•©
    text = f"""
ë ˆìŠ¤í† ë‘ ì´ë¦„: {row['restaurant_name']} ({row['restaurant_name_en']})
ìœ„ì¹˜: {row['place_en']}
ìŒì‹ ì¢…ë¥˜: {row['type_en']}
ì„¤ëª…: {row['description_clean_en']}
    """.strip()
    
    # 2. Document ìƒì„±
    doc = Document(
        page_content=text,
        metadata={
            # ID ë° ê²€ìƒ‰/í•„í„°ë§ì— ìì£¼ ì“°ì¼ í•µì‹¬ ì»¬ëŸ¼
            'restaurant_id': str(row['restaurant_id']),
            'restaurant_name_ko': str(row['restaurant_name']), # í•œê¸€ ë ˆìŠ¤í† ë‘ëª…
            'restaurant_name_en': str(row['restaurant_name_en']), # ì˜ë¬¸ ë ˆìŠ¤í† ë‘ëª…
            'place_en': str(row['place_en']), # ìœ„ì¹˜ (ì§€ì—­)
            'latitude': float(row['Latitude']), # LatitudeëŠ” ëŒ€ë¬¸ìë¡œ ì…ë ¥
            'longitude': float(row['Longitude']), # LongitudeëŠ” ëŒ€ë¬¸ìë¡œ ì…ë ¥
            
            # ëª¨ë“  ë‚˜ë¨¸ì§€ ì»¬ëŸ¼ (Payloadì— ì €ì¥ë¨)
            'image_path': str(row['image_path']),
            'near_subway_en': str(row['near_subway_en']),
            'type_en': str(row['type_en']),
            'description_clean_en': str(row['description_clean_en']),
        }
    )
    docs.append(doc)

print(f"âœ… {len(docs)}ê°œ Document ìƒì„± ì™„ë£Œ!")
print(f"\nì²« ë²ˆì§¸ í•­ëª© ìƒ˜í”Œ:")
print(docs[0].page_content)
```

```python
os.environ["OPENAI_API_KEY"] = "."  # â† ìˆ˜ì •!

# íŒŒì¼ ë° ì„¤ì •
CSV_FILE = "./entertainment_ver.csv"
QDRANT_URL = "http://localhost:6333"
COLLECTION_NAME = "seoul-restaurant"
                                                                                                                                                                                                              
print("âœ… ì„¤ì • ì™„ë£Œ!")
```

```python
print("ğŸ¤– OpenAI ì„ë² ë”© ëª¨ë¸ ì¤€ë¹„ ì¤‘...")

embedding = OpenAIEmbeddings(model="text-embedding-ada-002")

print("âœ… ì„ë² ë”© ëª¨ë¸ ì¤€ë¹„ ì™„ë£Œ!")
```

```python
# ì…€ 5: Qdrant ì—°ê²° ë° ì»¬ë ‰ì…˜ ìƒì„±
print("ğŸ”Œ Qdrant ì—°ê²° ì¤‘...")

qdrant_client = QdrantClient(url=QDRANT_URL)

# ê¸°ì¡´ ì»¬ë ‰ì…˜ ì‚­ì œ
try:
    qdrant_client.delete_collection(collection_name=COLLECTION_NAME)
    print("ğŸ—‘ï¸  ê¸°ì¡´ ì»¬ë ‰ì…˜ ì‚­ì œë¨")
except:
    print("â„¹ï¸  ê¸°ì¡´ ì»¬ë ‰ì…˜ ì—†ìŒ")

# ìƒˆ ì»¬ë ‰ì…˜ ìƒì„±
qdrant_client.create_collection(
    collection_name=COLLECTION_NAME,
    vectors_config=VectorParams(size=1536, distance=Distance.COSINE)
)

print(f"âœ… ì»¬ë ‰ì…˜ '{COLLECTION_NAME}' ìƒì„± ì™„ë£Œ!")
```

```python
# ì…€ 6: ë°ì´í„° ì„ë² ë”© ë° ì €ì¥
print(f"\nğŸ’¾ {len(docs)}ê°œ ë ˆìŠ¤í† ë‘ ì„ë² ë”© ìƒì„± ë° ì €ì¥ ì¤‘...")
print("â³ OpenAI API í˜¸ì¶œ ì¤‘... (ì•½ 10-15ë¶„ ì†Œìš”)")

vector_store = Qdrant.from_documents(
    documents=docs,
    embedding=embedding,
    url=QDRANT_URL,
    collection_name=COLLECTION_NAME,
)

print(f"\nâœ… ì´ {len(docs)}ê°œ ë ˆìŠ¤í† ë‘ ì €ì¥ ì™„ë£Œ!")
print(f"ğŸ“Š Qdrant UI: {QDRANT_URL}/dashboard")
```
